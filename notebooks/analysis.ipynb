```python
# Here I start checking for missing values in dataset 

missing_values = df.isnull().sum()

print(missing_values )
```

```python
# Calculate the total number of cells with missing values

total_missing_values = missing_values.sum()

print("Total Missing value:", total_missing_values)
```

```python
# Here I calculate the total cells in dataset

total_columns = df.shape[1]

total_rows = df.shape[0]

total_cells = total_rows * total_columns 

print(total_cells)
```

```python
overall_missing_percentage = ( total_missing_values / total_cells ) * 100

print(f"Overall Missing Percentage: {overall_missing_percentage:.2f}%")
```

```python
# This accounts for the percentage of data missing from the columns with missing values

print(missing_values["location_northing_osgr"] / total_rows * 100)

print(missing_values["location_easting_osgr"] / total_rows * 100)
```
